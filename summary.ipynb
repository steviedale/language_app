```python
import os
import pandas as pd
```

```python
lessons_df = pd.DataFrame()
lesson_dir = 'language_notes/spotify_lessons'
for lesson in  os.listdir(lesson_dir):
    path = os.path.join(lesson_dir, lesson, 'transcript.csv')
    assert(os.path.exists(path))
    df = pd.read_csv(path)
    lessons_df = pd.concat([lessons_df, df], ignore_index=True)
```

```python
len(lessons_df)
```

```python
italian_phrases = lessons_df['italian']
len(italian_phrases)
```

```python
italian_words = set(lessons_df['italian'].apply(lambda x: x.split(' ')).explode())
italian_words = [x.lower() for x in italian_words if len(x) > 0 and x != '/']
italian_words = sorted(italian_words)
len(italian_words)
```
